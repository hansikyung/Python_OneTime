## 5. 연산자    
- 할당연산자
- 비교연산자
- 논리연산자
- 멤버연산자
- 식별연산자

```python
#할당연산자
A=1
A=A+1
print(A)
```

```python
B=1
B+=1
print(B)
```

```python
#비교연산자
A=1
B=32
print(A>B)
```

```python
C=-3
D=-0.2
print(C<=D)
```

```python
#==, !=
A=1
B=32
print(A==B)
```

```python
#논리연산자
A = True
B = False

print(A and B)
print(A or B)
print(not A)
print(not B)
```

```python
#멤버연산자
A = '사과'
B = ['사과', '배', '참외', '수박', '복숭아']
print(A in B)
```

```python
C = 'apple'
D = ['Apple', 'pear', 'watermelon', 'peach']
print(C in D)
```

```python
#식별연산자
A = '사과'
B = '사과'
print(A is B)
```

```python
C = 12
D = 12
print(C is not D)
```

## 문자열 다루기

```python
#문자 양 옆에 공백이 있으니 주의하세요!
comment = ' I love programming. '
```

```python
print(comment.lstrip())
print(comment.rstrip())
print(comment.strip())
```

```python
print(comment.replace('.', '!!!!'))
```

```python
print(comment.split(' '))
```

```python
print(comment.count('i'))
```

```python
print('-'.join(comment))
```
