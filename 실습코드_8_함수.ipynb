## 1. 함수

```python
# 매개변수가 없는 함수 정의
def greet():
    print("안녕하세요!")

# 함수 호출
greet()  # 출력: 안녕하세요!
```

```python
# 매개변수가 있는 함수 정의
def greet_with_name(name):
    print(f"안녕하세요, {name}님!")

# 함수 호출
greet_with_name("철수")  # 출력: 안녕하세요, 대리님!
greet_with_name("영희")  # 출력: 안녕하세요, 사장님!
```

```python
# 두 수를 더하는 일반 함수 정의
def add(x, y):
    return x + y

# 함수 호출
result = add(5, 3)
print("일반 함수 add(5, 3)의 결과:", result)
```

## 추가 함수 예제

```python
# 일반 함수로 요일 판별 함수 정의
def day_of_week(day_number):
    days = ["일요일", "월요일", "화요일", "수요일", "목요일", "금요일", "토요일"]
    if 0 <= day_number <= 6:
        return days[day_number]
    else:
        return "유효하지 않은 숫자입니다."

# 함수 호출
print(day_of_week(0))  # 출력: 일요일
print(day_of_week(3))  # 출력: 수요일
print(day_of_week(7))  # 출력: 유효하지 않은 숫자입니다.
```
