##4. 셋의 값 추가 제거

```python
S1 = set([1, 3, 5, 7, 9, 10])
S2 = set([2, 4, 6, 8, 10])
```

```python
#값 하나 추가
S2.add(12)
print(S2)
```

```python
#여러 개의 값 추가
S1.update([11, 13, 15])
print(S1)
```

```python
#특정 값 제거
S1.remove(10)
print(S1)
```

## 5. 셋의 교집합, 차집합, 합집합

```python
S1 = set([1, 3, 5, 7, 9, 10])
S2 = set([2, 4, 6, 8, 10])
```

```python
print(S1&S2)
```

```python
print(S1.intersection(S2))
```

```python
print(S1-S2)
```

```python
print(S1.difference(S2))
```

```python
print(S1|S2)
```

```python
print(S1.union(S2))
```
