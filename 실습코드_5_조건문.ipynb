```python
# 주민등록번호를 입력받아 성별을 판별하는 코드
id_number = input("주민등록번호를 입력하세요 (예: 901010-1234567): ")

# 주민등록번호의 7번째 숫자를 확인
gender_digit = id_number[7]  # 7번째 문자가 성별을 나타냄

# 성별 판별
if gender_digit == '1' or gender_digit == '3':
    print("남자입니다.")
elif gender_digit == '2' or gender_digit == '4':
    print("여자입니다.")
else:
    print("올바른 주민등록번호를 입력하세요.")
```
