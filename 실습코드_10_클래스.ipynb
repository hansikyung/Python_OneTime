## 1. 클래스의 생성

```python
class fish_pastry:
    def __init__(self, flour : int, filling):
      self.flour
      self.filling

    def bake():
      print('안녕 나는 붕어빵!')
```

```python
# fish_pastry 클래스 정의
class fish_pastry:
    def __init__(self, flour : int, filling):
        # __init__ 메서드에서 밀가루(flour)와 앙금(filling)을 초기화
        self.flour = flour
        self.filling = filling

    # bake 메서드: n 마리 완성! 이라고 인쇄
    def bake(self, n):
        if self.flour < n*50:
          print(f"밀가루가 부족하니 최대 {n*50}의 밀가루를 투입하세요.")
        else:
          self.flour = self.flour - n*50
          print(f"{n}마리 완성!")

    # get_item 메서드: 앙금을 반환
    def get_item(self):
        return self.filling
```

```python
# fish_pastry 클래스 사용
# 붕어빵 객체 생성 (밀가루: '밀가루 1kg', 앙금: '팥앙금')
my_pastry = fish_pastry(500, "팥앙금")

# bake 메서드 호출
my_pastry.bake(10)  # 출력: 10마리 완성!

# get_item 메서드 호출
filling = my_pastry.get_item()
print("이 붕어빵의 앙금은:", filling)  # 출력: 이 붕어빵의 앙금은: 팥앙금
```

```python
# bake 메서드 호출
my_pastry.bake(1)  # 출력: 10마리 완성!
```

## 2. 상속 : 이제 붕어빵 말고 땅콩빵을 팔기로 했거든요?

코드 재사용: 부모 클래스에 이미 정의된 메서드와 속성을 자식 클래스에서 다시 정의할 필요 없이 사용할 수 있습니다.         
유지보수 용이: 공통된 기능을 부모 클래스에 모아두면, 자식 클래스들은 해당 기능을 그대로 상속받아 사용할 수 있습니다.          
확장성: 자식 클래스에서 부모 클래스의 기능을 확장하거나 수정할 수 있습니다.

```python
# 자식 클래스 peanut_bread 정의 (fish_pastry 상속)
class peanut_bread(fish_pastry):
    def __init__(self, flour, filling, peanuts):
        # 부모 클래스의 __init__을 호출하여 밀가루와 앙금을 초기화
        super().__init__(flour, filling)
        self.peanuts = peanuts  # 땅콩 추가

    # 땅콩을 반환하는 메서드
    def get_peanuts(self):
        return self.peanuts
```

```python
# peanut_bread 클래스 사용 예시
my_peanut_bread = peanut_bread(250, "팥앙금", "땅콩 50g")

# bake 메서드 호출
my_peanut_bread.bake(5)  # 출력: 5마리 완성!

# get_item 메서드 호출 (앙금 반환)
print("앙금:", my_peanut_bread.get_item())  # 출력: 앙금: 팥앙금

# get_peanuts 메서드 호출 (땅콩 반환)
print("땅콩:", my_peanut_bread.get_peanuts())  # 출력: 땅콩: 땅콩 50g
```

###### 아 근데 5개가 아니라 5마리가 완성이네? 뭔가 이상

```python
# 자식 클래스 peanut_bread 정의 (fish_pastry 상속)
class peanut_bread(fish_pastry):
    def __init__(self, flour, filling, peanuts):
        super().__init__(flour, filling)
        self.peanuts = peanuts

    def get_peanuts(self):
        return self.peanuts

    # bake 메서드 오버라이딩
    def bake(self, n):
        # 부모 클래스의 bake 메서드와는 다른 동작을 정의
        if self.flour < n*30:
          print(f"밀가루가 부족하니 최대 {n*30}의 밀가루를 투입하세요.")
        else:
          self.flour = self.flour - n*30
          print(f"{n}개의 땅콩빵 완성! 땅콩 포함!")
```

```python
# peanut_bread 클래스 사용 예시
my_peanut_bread = peanut_bread(150, "팥앙금", "땅콩 50g")

# 오버라이딩된 bake 메서드 호출
my_peanut_bread.bake(5)  # 출력: 5개의 땅콩 붕어빵 완성! 땅콩 포함!

# 부모 클래스의 메서드인 get_item 호출
print("앙금:", my_peanut_bread.get_item())  # 출력: 앙금: 팥앙금

# get_peanuts 메서드 호출
print("땅콩:", my_peanut_bread.get_peanuts())  # 출력: 땅콩: 땅콩 50g
```

## 다형성


```python
class Animal:
    def speak(self):
        raise NotImplementedError("이 메서드는 자식 클래스에서 구현해야 합니다.")

class Dog(Animal):
    def speak(self):
        return "멍멍"

class Cat(Animal):
    def speak(self):
        return "야옹"

animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())  # 출력: 멍멍, 야옹
```
