## 1. 람다식

```python
# 두 수를 더하는 람다식
add_lambda = lambda x, y: x + y

# 람다식 호출
result_lambda = add_lambda(5, 3)
print("람다식 add_lambda(5, 3)의 결과:", result_lambda)
```

```python
# 숫자가 짝수인지 홀수인지 판별하는 람다식
even_or_odd = lambda x: "짝수" if x % 2 == 0 else "홀수"

# 람다식 호출
print(even_or_odd(4))  # 출력: 짝수
print(even_or_odd(7))  # 출력: 홀수
```

```python
# 숫자가 양수, 음수, 0인지 판별하는 람다식
number_type = lambda x: "양수" if x > 0 else ("음수" if x < 0 else "0")

# 람다식 호출
print(number_type(10))  # 출력: 양수
print(number_type(-5))  # 출력: 음수
print(number_type(0))   # 출력: 0
```

## 2. 함수와 람다식 비교하기

```python
# 정렬 기준으로 사용할 함수 정의
def sort_by_second_element(t):
    return t[1]

# 리스트와 일반 함수로 정렬
my_list = [(1, 5), (2, 3), (3, 8)]
sorted_list = sorted(my_list, key=sort_by_second_element)
print("일반 함수로 정렬된 리스트:", sorted_list)
```

```python
# 람다식을 사용하여 정렬
my_list = [(1, 5), (2, 3), (3, 8)]
sorted_list_lambda = sorted(my_list, key=lambda t: t[1])
print("람다식으로 정렬된 리스트:", sorted_list_lambda)
```

##### 다른 함수


```python
# 일반 함수로 요일 판별 함수 정의
def day_of_week(day_number):
    days = ["일요일", "월요일", "화요일", "수요일", "목요일", "금요일", "토요일"]
    if 0 <= day_number <= 6:
        return days[day_number]
    else:
        return "유효하지 않은 숫자입니다."

# 함수 호출
print(day_of_week(0))  # 출력: 일요일
print(day_of_week(3))  # 출력: 수요일
print(day_of_week(7))  # 출력: 유효하지 않은 숫자입니다.
```

```python
# 람다식으로 요일 판별
day_of_week_lambda = lambda day_number: ["일요일", "월요일", "화요일", "수요일", "목요일", "금요일", "토요일"][day_number] if 0 <= day_number <= 6 else "유효하지 않은 숫자입니다."

# 람다식 호출
print(day_of_week_lambda(0))  # 출력: 일요일
print(day_of_week_lambda(3))  # 출력: 수요일
print(day_of_week_lambda(7))  # 출력: 유효하지 않은 숫자입니다.
```

## **. map

```python
# 리스트의 각 요소에 2를 곱하는 함수
def multiply_by_two(x):
    return x * 2

# 리스트에 map() 함수 적용
numbers = [1, 2, 3, 4, 5]
result = list(map(multiply_by_two, numbers))
print("map() 함수 결과:", result)

# 람다식을 사용하여 같은 작업을 수행
result_lambda = list(map(lambda x: x * 2, numbers))
print("map() 함수와 람다식 결과:", result_lambda)
```
