## 2. 리스트의 추가, 삭제


```python
listA.append('새로운 데이터')
print(listA)
```

```python
listA.insert(1, 2)
print(listA)
```

```python
listA.remove('새로운 데이터')
print(listA)
```

```python
del listA[0]
print(listA)
```
