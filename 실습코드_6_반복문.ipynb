## 반복문

```python
for a in [1, 2, 3, 4, 5]:
  print(a)
```

```python
for i in range(10):
  print(i)
```

```python
count=['가', '나', '다', '라']
for i in range(len(count)):
  print(i)
```

```python
dict_sample = {1:'candy', 2:'ice cream', 3:4}
for i, v in enumerate(dict_sample):
  print(i , '--->' , v)
```

```python
count=['가', '나', '다', '라']
numbers = [1, 2, 3, 4]

for i, v in zip(count, numbers):
  print(i, ":", v)
```
